```python
from bs4 import BeautifulSoup as bs
import requests as r
import pandas as pd
import numpy as np
```

```python
# Function to extract the Product title
def get_title(soup):
    try:
        # Our Tag object
        title = soup.find("span", class_="B_NuCI")
        # Inner NavigatbleString Object
        title_value = title.text
        # if you can see the gap b/w Title as a String Value
        title_string = title_value.strip()
    except AttributeError:
        title_string = ""
    return title_string

# Function to extract the Product Price
def get_price(soup):
    try:
        price = soup.find("div", class_="CEmiEU").find('div', attrs={'class': '_30jeq3 _16Jk6d'}).text
    except AttributeError:
        try:
            # if there is a dealprice
            price = soup.find("div", class_="CEmiEU").find('div', attrs={'class': '_3I9_wc _2p6lqe'}).text
        except:
            price = " "
    return price

def get_rating(soup):
    try:
        rating = soup.find("div", class_="_3LWZlK").text
    except AttributeError:
        # try:
        #     # if there is a dealprice
        #     rating = soup.find("div", class_="CEmiEU").find('div', attrs={'class': '_2_R_DZ'}).text
        # except:
        rating = " "
    return rating

def get_review(soup):
    try:
        review_count = soup.find("div", class_="_2_R_DZ").text
    except AttributeError:
        review_count = " "
    return review_count

def get_availability(soup):
    try:
        available = soup.find("div", class_="_3TT44I").text
    except AttributeError:
        available = "Not Available "
    return available
```

```python
import requests as r
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np

# Rest of your code...

if __name__ == '__main__':
    HEADERS = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36',
        'Accept-Language': 'en-US, en;q=0.5',
    }
    URL = "https://www.flipkart.com/search?q=gaming&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"

    webpage = r.get(URL, headers=HEADERS)

    soup = bs(webpage.content, 'html.parser')
    links = soup.find_all('a', attrs={'class': 's1Q9rs'})

    link_list = []
    for link in links:
        link_list.append(link.get('href'))

    d = {'title': [], 'price': [], 'rating': [], 'review': [], 'availability': []}
    for link in link_list:
        new_webpage = r.get("https://www.flipkart.com" + link, headers=HEADERS)
        new_soup = bs(new_webpage.content, 'lxml')
        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['review'].append(get_review(new_soup))
        d['availability'].append(get_availability(new_soup))

    Flipkart_df = pd.DataFrame.from_dict(d)
    Flipkart_df['title'].replace(' ', np.nan, inplace=True)
    Flipkart_df = Flipkart_df.dropna(subset=['title'])
    Flipkart_df.to_csv('Flipkart_data.csv', header=True, index=False)
```
